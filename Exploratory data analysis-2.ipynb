# FEATURE ENGINEERING

```python
# Feature Engineering
# import the libraries
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv('housing_data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
# Feature Selection
selected_features = ['LotArea','YearBuilt']
data_selected = data[selected_features]
```

```python
data_selected
```

```python
# Feature Scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[['LotArea','YearBuilt']])
```

```python
scaled_data
```

```python
# Feature Transformation
data_transformed = np.log1p(data[['YearBuilt']])
```

```python
data_transformed
```

```python
# Handling missing Values
imputer = SimpleImputer(strategy='mean')
data_imputed = imputer.fit_transform(data[['YearBuilt']])
```

```python
data_imputed 
```

```python
# Encoding the categorical Variables
encoder = OneHotEncoder()
data_encoded = encoder.fit_transform(data[['YearBuilt']])
```

```python
print(data_encoded)
```

```python
# Creating Interaction Terms
poly = PolynomialFeatures(degree=2, interaction_only=True)
data_interactions = poly.fit_transform(data[['YearBuilt']])
```

```python
data_interactions
```

```python
# Feature Aggregation
data_aggregated = data.groupby('LotArea').agg({'YearBuilt': 'mean', 'YearRemodAdd': 'count'})
```

```python
data_aggregated
```

```python
scaled_data
```

```python
scaled_data = scaler.fit_transform(data_imputed)
```

```python
# Dimensionality Reduction
pca = PCA(n_components = 1)
data_pca = pca.fit_transform(scaled_data)
```
